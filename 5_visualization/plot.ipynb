```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for the bar chart
data = {
    'Feature Extraction Method': ['TFIDF', 'WORD2VEC', 'BERT', 'SENTENCE_BERT', 'SENTENCE_BERT+MLM'],
    'C_v': [0.5088, 0.6191, 0.6214, 0.6848, 0.6956],
    'C_pmi': [-0.0187, 0.0851, 0.0359, 0.1037, 0.1047],
    'C_uci': [-0.1764, 0.4442, 0.1412, 0.5975, 0.5726]
}

df = pd.DataFrame(data)

# Plotting the bar chart
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.2
bar_positions = range(len(df))

# Create bars for each metric
bar1 = ax.bar([p - bar_width for p in bar_positions], df['C_v'], width=bar_width, label='C_v')
bar2 = ax.bar(bar_positions, df['C_pmi'], width=bar_width, label='C_pmi')
bar3 = ax.bar([p + bar_width for p in bar_positions], df['C_uci'], width=bar_width, label='C_uci')

# Adding labels and title
ax.set_xlabel('Feature Extraction Method')
ax.set_ylabel('Scores')
ax.set_title('Scores by Feature Extraction Method and Metric')
ax.set_xticks(bar_positions)
ax.set_xticklabels(df['Feature Extraction Method'])
ax.legend()

# Adding the data labels on top of the bars
for bars in [bar1, bar2, bar3]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for the bar chart
data = {
    'Feature Extraction Method': ['TFIDF', 'WORD2VEC', 'BERT', 'SENTENCEBERT', 'SENTENCEBERT+MLM'],
    'Jaccard': [1.0, 0.0386, 0.0361, 0.0321, 0.0216],
    'CosineSim': [0.5474, 0.7123, 0.7150, 0.7352, 0.6967]
}

df = pd.DataFrame(data)

# Plotting the bar chart
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
bar_positions = range(len(df))

# Create bars for each metric
bar1 = ax.bar([p - bar_width/2 for p in bar_positions], df['Jaccard'], width=bar_width, label='Jaccard')
bar2 = ax.bar([p + bar_width/2 for p in bar_positions], df['CosineSim'], width=bar_width, label='CosineSim')

# Adding labels and title
ax.set_xlabel('Feature Extraction Method')
ax.set_ylabel('Scores')
ax.set_title('Scores by Feature Extraction Method and Metric')
ax.set_xticks(bar_positions)
ax.set_xticklabels(df['Feature Extraction Method'])
ax.legend()

# Adding the data labels on top of the bars
for bars in [bar1, bar2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.show()
```
